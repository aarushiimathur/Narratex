### Audio Generation

```python
def text_to_speech(text):
    try:
        print("Generating speech...")

        words = text.split()  # Split the text into words
        combined_audio = AudioSegment.silent(duration=0)  # Start with empty audio

        for word in words:
            # Generate speech for each word
            tts = gTTS(text=word, lang='en')
            word_audio_path = f"{word}.mp3"
            tts.save(word_audio_path)
            #print(f"Audio saved for word '{word}' as {word_audio_path}")

            # Load the word audio file
            word_audio = AudioSegment.from_mp3(word_audio_path)

            # Add the word's audio to the combined audio
            combined_audio += word_audio

            # Add a 2-second silence after each word
            silence = AudioSegment.silent(duration=2000)  # 2000 ms = 2 seconds
            combined_audio += silence

            # Clean up the temporary word audio file
            os.remove(word_audio_path)

        # Save the final combined audio
        combined_audio.export("output_with_gaps.mp3", format="mp3")
        print("Audio with gaps saved as output_with_gaps.mp3")
        os.system(f"mpg321 output_with_gaps.mp3")

    except Exception as e:
        print("Error in text-to-speech:", e)
    return words, combined_audio
```

### Data Extraction

```python
#Function to extract images from PDF
def extract_images_from_pdf(pdf_path):
    images = convert_from_path(pdf_path)
    return images  # Returns a list of PIL images
```

```python
# Function to extract text from an image
def extract_text_from_image(image):
    return pytesseract.image_to_string(image)
```

```python
# Function to detect type of content (numbers, alphabets, words)
def extract_sequences(text):
    lines = text.split('\n')

    # Find the first line with meaningful content (ignoring instructions)
    for i, line in enumerate(lines):
        if re.search(r'\d{5,}', line) or re.search(r'^[A-Z](?: [A-Z])+$', line, re.I) or re.search(r'^\b[a-zA-Z]{3,}\b(?: \b[a-zA-Z]{3,}\b)+$', line):
            sequence_lines = lines[i:]  # Capture everything after instructions
            break
    else:
        return "", "unknown"

    # Extract sequences
    numbers = re.findall(r'\d', ' '.join(sequence_lines))
    alphabets = re.findall(r'^[A-Z]', ' '.join(sequence_lines), re.I)
    words = re.findall(r'\b[a-zA-Z]{3,}', ' '.join(sequence_lines))

    # Determine the type of sequence
    if len(numbers) > 0 and len(numbers) > len(alphabets) and len(numbers) > len(words):
        return ' '.join(numbers), "numbers"
    elif len(alphabets) > 0 and len(0+0+0+0) > len(words):
        return ' '.join(alphabets), "alphabets"
    elif len(words) > 0:
        return ' '.join(words), "words"
    else:
        return "", "unknown"
```

```python
# Function to process a PDF containing images
def process_pdf(file_path):
        try:
            images = extract_images_from_pdf(file_path)  # Convert PDF pages to images
            all_text = ""

            for img in images:
                text = extract_text_from_image(img)  # Extract text from each image
                all_text += text + "\n"

            sequences, sequence_type = extract_sequences(all_text)

            if sequences:
                print(f"Extracted {sequence_type.capitalize()}:\n", sequences)
                text_to_speech(sequences)
                print("Audio generated")
            else:
                print("No valid sequence found.")
        except Exception as e:
            print(f"Error processing PDF: {e}")
```

```python
def extract_story_from_txt(file_path):
    """Extracts the story text from a text file."""
    with open(file_path, "r", encoding="utf-8") as file:
        story_text = file.read()
    return story_text.strip()
```

```python
# Prosody evaluation
import librosa
import numpy as np
import matplotlib.pyplot as plt

def analyze_f0(audio_path):
    # Load audio
    signal, sr = librosa.load(audio_path)

    # Ensure signal has meaningful amplitude
    if np.max(signal) == 0:
        print("Audio signal is silent or flat.")
        return

    # Calculate F0 using librosa.pyin
    f0, voiced_flag, voiced_probs = librosa.pyin(
        signal,
        fmin=librosa.note_to_hz('C2'),
        fmax=librosa.note_to_hz('C6'),
        sr=sr
    )

    # Handle missing F0 values (None)
    f0_cleaned = np.nan_to_num(f0)  # Replace None with 0

    # Plot F0 values
    plt.figure(figsize=(10, 4))
    plt.plot(f0_cleaned, label='F0 (Hz)')
    plt.xlabel('Time Frame')
    plt.ylabel('F0 (Hz)')
    plt.title('Fundamental Frequency Over Time')
    plt.legend()
    plt.show()

audio_path = "/content/output_with_gaps.mp3"
analyze_f0(audio_path)
```

 Prosody evaluation is how speech conveys meaning, emotion, and emphasis. The ideal graph should contain variations.

In our audio, we need no variation in the graph since our speech needs to be monotonous.
